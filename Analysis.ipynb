```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('results.csv')
df.head(10)
```

```python
ax = sns.relplot(x='Hour',y='Max_High',data=df,kind='line',hue='Stock_Name')
ax.fig.set_size_inches(12,10)
```

# Graph 1
According to this graph, each stock had a very pretty stable trend on the May 14, 2020. 
None of these stocks increased or decreased dramtariclly throughout the day.

```python
ax2 = sns.catplot(x='Stock_Name',y='Max_High',data=df, kind='bar')
ax2.fig.set_size_inches(12,10)
```

# Graph 2
This graph is a bar plot to displays mean of max high price per stock. Since each stock has pretty stable trend, the 95% condence intervals for the mean per each stock is not big.

```python
fig = plt.figure(figsize=(12,10))
df[df.Stock_Name.isin(['NFLX'])].groupby(['Stock_Name'])['Max_High'].plot() 
plt.xlabel('Hour')
plt.ylabel('Highest_Price')
plt.xticks([24, 25, 26, 27, 28,29,30], [9,10,11,12,13,14,15])
```

# Graph 3
This graph shows the stock price change of Netflix from each hour. On May 14, 2020, Netflix dropped about 10 dollars starting from 9 AM to 11 AM. Then, it slowly increased, but by the end of the day it lost 10 dollars per stock. 

```python
fig = plt.figure(figsize=(12,10))
df[df.Stock_Name.isin(['FB'])].groupby(['Stock_Name'])['Max_High'].plot() 
plt.xlabel('Hour')
plt.ylabel('Highest_Price')
plt.xticks([17, 18, 19, 20, 21, 22, 23], [9,10,11,12,13,14,15])
```

# Graph4
This graph shows the stock price change of Facebook from each hour. On May 14, 2020, the change of Facebook is not much, but if we take a close look, Facebook increased 3 dollars per stock by the end of the day.
